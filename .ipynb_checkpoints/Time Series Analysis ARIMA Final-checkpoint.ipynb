```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Set Seaborn style
sns.set(style="whitegrid")

# Set the figure size
plt.rcParams['figure.figsize'] = (16, 10)

# Perform seasonal decomposition
decomposition = seasonal_decompose(df_week, period=52, model='additive')

# Plot the original time series
plt.subplot(4, 1, 1)
plt.plot(decomposition.observed, color='blue', label='Original')
plt.legend(loc='upper left')
plt.title('Original Time Series')

# Plot the trend component
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, color='green', label='Trend')
plt.legend(loc='upper left')
plt.title('Trend Component')

# Plot the seasonal component
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, color='red', label='Seasonal')
plt.legend(loc='upper left')
plt.title('Seasonal Component')

# Plot the residual component
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, color='purple', label='Residual')
plt.legend(loc='upper left')
plt.title('Residual Component')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

```

#### The seasonal_decompose function is part of the statsmodels library in Python and is used for time series decomposition. It decomposes a time series into several components, allowing you to analyze and understand the underlying patterns. The components obtained from seasonal decomposition are:

*Observed (or Original) Series:

-  This is the original time series data that we input to the seasonal_decompose function.

*Trend Component:

- The trend component represents the long-term progression or movement in the data.
- It captures the overall direction of the series, ignoring short-term fluctuations and noise.

*Seasonal Component:

- The seasonal component represents repeating weekly  patterns or cycles within the data.

*Residual (or Irregular) Component:

- The residual component represents the remaining random fluctuations or noise in the data.
- It includes everything that cannot be attributed to the trend or the seasonal component.

## Data Transformation

**_The adf_check() function performs an Augmented Dickey-Fuller test on a time series. The test is used to determine whether a time series is stationary or not. If the p-value of the test is less than or equal to 0.05, then there is strong evidence against the null hypothesis, and the series is considered to be stationary. Otherwise, the series is considered to be non-stationary. The function takes a time series as input and returns the results of the test. The results are printed to the console, along with a message indicating whether the series is stationary or not._**

```python
def adf_check(time_series):
    result = adfuller(time_series , autolag = 'AIC')
    label = pd.Series(result[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in result[4].items():
        label['Critical Value (%s)'%key] = value
    print(label)
    if result[1] <= 0.05:
        print('Strong evidence against the null hypothesis, hence REJECT null hypothesis and the series is Stationary')
    else:
        print ('Weak evidence against the null hypothesis, hence ACCEPT null hypothesis and the series is Not Stationary ')
```

**Performing the Augmented Dickey Fuller Test on the original data in the  resampled weekly dataset to find whether the time series is a stationary or not**

```python
adf_check(df_week)
```

# Model Training and Fitting


## Model 1 : ARIMA (AutoRegressive Integrated Moving Average)

####  We chosed  ARIMA model which is popular and frequently used for time-series analysis and forecasting.auto_arima is a function that automates the process of selecting the optimal p, d, and q values for an ARIMA model. It uses a stepwise approach and considers different combinations of parameters to find the model with the lowest Akaike Information Criterion (AIC) or another specified criterion.**Getting the p value and q value for the model fitting using `auto_arima` function by passing through some needed parameters , the best model is evaluated by least Akaike Information Criterion (AIC)**

```python
model = auto_arima(df_week , m = 52, d = 1 ,seasonal = False , max_order = 8 , test = 'adf' , trace = True)
```

* df_week: The time series data for which we want to find the best ARIMA model.

* m=52: The seasonality of the time series. In this case, it's set to 52, which is a weekly seasonality.

* d=1: The order of differencing. It indicates how many times the differencing operation is applied to make the time series stationary.

* seasonal=False: This parameter is set to False since we've specified the seasonality separately using the m parameter.

* max_order=8: This parameter specifies the maximum value for p, d, and q in the ARIMA model. In this case, it's set to 8.

* test='adf': The type of unit root test to use for stationarity testing. 'adf' stands for Augmented Dickey-Fuller, a commonly used test for stationarity.

* trace=True: This parameter prints the detailed debugging output during the fitting process.

* The auto_arima function works by automatically selecting the optimal combination of p, d, and q parameters for the ARIMA model based on minimizing the Akaike Information Criterion (AIC). It performs a search over a range of possible model configurations and chooses the one that yields the lowest AIC. This helps in automating the process of finding the best ARIMA model without the need for manual tuning.

* The AIC is a measure of the model's goodness of fit, penalized by the number of parameters. Lower AIC values indicate a better-fitting model. The auto_arima function evaluates multiple models with different combinations of p, d, and q and selects the one with the lowest AIC, making it a convenient tool for automatic model selection in time series analysis.

* p: The order of the autoregressive part.
* d: The order of differencing.
* q: The order of the moving average part.

## From the above model :
### ARIMA(p, d, q):
- p (AR order): 0 - There are no autoregressive (AR) terms in the model.
- d (Differencing order): 1 - The time series is differenced once to achieve stationarity.
- q (MA order): 1 - There is one moving average (MA) term in the model.

### (P, D, Q)[s]:
- P (Seasonal AR order): 0 - There are no seasonal autoregressive (SAR) terms in the model.
- D (Seasonal differencing order): 0 - There is no seasonal differencing.
- Q (Seasonal MA order): 0 - There are no seasonal moving average (SMA) terms in the model.
- s (Seasonal period): 0 - There is no seasonality in this model.

* So, the best model ARIMA(0,1,1)(0,0,0)[0] is a non-seasonal ARIMA model with one differencing term and one moving average term. It has no seasonal components.

**_Defining the summary of the model fitted with auto_arima function , here getting various information such as Akaike Information Criterion (AIC), Bayesian Information Criterion (BIC) , Hannan-Quinn Information Criterion (HQIC), Log Likelihood etc. From which we can evaluate the model_**

## Model Testing
**_Splliting the Resampled Weekly Dataset into training data and test data. Displaying the number of observations in both training data and test data._**

```python
size = int(len(df_week)*0.80)
train , test = df_week[0:size]['Value'] , df_week[size:(len(df_week))]['Value']
print('Counts of Train Data : ',train.shape[0])
print('Counts of Train Data : ',test.shape[0])
```

**_Creating the list of train dataset values in train_values and empty predictions list which will be appended after the prediction. Then fitting the model with ARIMA model with the best value got from auto_arima model in the train_values data and predicting with test data value and appending it to the predictions list and printing the comparison between predicted value and actual value_**

```python
train_values = [x for x in train]
prediction = []
print('Printing Predictied vs Expected Values....')
print('\n')
for t in range(len(test)):
    model = ARIMA(train_values , order = (0,1,1))
    model_fit = model.fit()
    output = model_fit.forecast()
    pred_out = output[0]
    prediction.append(float(pred_out))
    test_in = test[t]
    train_values.append(test_in)
    print('Predicted = %f , Actual = %f' % (pred_out , test_in))
```

**_Evaluating the model with the following metrics such as percentage of r2 score , mean squared error , root mean squared error , mean absolute error and mean absolute percentage error for the predicted value and test data._**

```python
print('Evaluation Result for Test data : ','\n')
print('R2 Score for Test data : {0:.2f} %'.format(100*r2_score(test,prediction)),'\n')
print('Mean Squared Error : ',mean_squared_error(test,prediction),'\n')
print('Mean Absolute Error : ',mean_absolute_error(test,prediction),'\n')
print('Root Mean Squared Error : ',sqrt(mean_squared_error(test,prediction)),'\n')
print('Mean Absolute Percentage Error : {0:.2f} %'.format(100*mean_absolute_percentage_error(test,prediction)))
```

## Model 2 : LSTM

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


# Extract values and scale them
values = df_week['Value'].values.reshape(-1, 1)
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(values)

# Create sequences for training
def create_sequences(data, seq_length):
    sequences, labels = [], []
    for i in range(len(data) - seq_length):
        seq = data[i:i+seq_length, 0]  
        label = data[i+seq_length, 0]  
        sequences.append(seq)
        labels.append(label)
    return np.array(sequences), np.array(labels)

seq_length = 10
X_train, y_train = create_sequences(scaled_values, seq_length)

# Reshape input data for LSTM
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# Create and train LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(seq_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=50, batch_size=32)

# Create sequences for testing
test_size = len(scaled_values) - seq_length
X_test, y_test = create_sequences(scaled_values[-test_size:], seq_length)

# Reshape input data for LSTM
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions to the original scale
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

```

```python
# Plotting function
def plot_lstm_forecast(actual, forecast, title):
    plt.figure(figsize=(18, 12))
    plt.plot(actual, label='Actual', marker='o')
    plt.plot(forecast, label='Forecast', linestyle='dashed', color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
    
#plot Outcomes
plot_lstm_forecast(values[-test_size:], predictions, 'LSTM Forecast')
```
